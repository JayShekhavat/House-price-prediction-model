##Rename the column Price (INR) to Price

```python
data.columns
```

```python
data.rename(columns={'bedRoom': 'Bedroom', 'bathroom': 'Bathroom', 'balcony': 'Balcony','floorNum': 'Floors','area': 'Area','price': 'Price'}, inplace=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['Floors'] = data['Floors'].fillna(data['Floors'].mean())
```

```python
data['Area'] = data['Area'].fillna(data['Area'].mean())
```

```python
data['Price'] = data['Price'].fillna(data['Price'].mean())
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
print(type(data['Balcony']))
```

```python
data.Balcony.unique()
```

```python
data['Balcony'] = ['3' if item == '3+' else item for item in list(data['Balcony'])]
```

```python
data['Balcony']
```

```python
data.head()
```

```python
data['Balcony'] = data['Balcony'].astype(int)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

#Apply algorithm to make model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
random_forrest_model = RandomForestRegressor(
    n_estimators=100,  # Number of trees
    max_depth=10,      # Maximum depth of the tree
    random_state=42    # For reproducibility
)
```

```python
random_forrest_model.fit(X_train, y_train)

# Make predictions
y_pred = random_forrest_model.predict(X_test)
```
